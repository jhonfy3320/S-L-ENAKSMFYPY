- El algoritmo Mean-Shift es un algoritmo de agrupamiento que busca automáticamente el número de clústeres en un conjunto de datos y asigna los puntos de datos a los clústeres en función de la densidad de puntos. El algoritmo funciona moviendo de manera iterativa los centroides de los clústeres hacia las regiones de mayor densidad de puntos hasta que converja a los máximos locales.

#### A continuación, te mostraré cómo implementar el algoritmo Mean-Shift en Python utilizando la biblioteca scikit-learn:

```python
import numpy as np
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generar datos de ejemplo
n_samples = 1000
n_features = 2
n_clusters = 3
X, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)

# Crear un modelo Mean-Shift
mean_shift = MeanShift()

# Ajustar el modelo a los datos
mean_shift.fit(X)

# Obtener las etiquetas de los clústeres
labels = mean_shift.labels_

# Obtener las coordenadas de los centroides
cluster_centers = mean_shift.cluster_centers_

# Número estimado de clústeres
n_clusters_estimado = len(cluster_centers)

# Graficar los datos y los clústeres
colors = plt.cm.Spectral(np.linspace(0, 1, n_clusters_estimado))
for i in range(n_clusters_estimado):
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[i], label=f'Cluster {i + 1}')

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=200, linewidths=3, color='black', label='Centroides')
plt.legend()
plt.title(f'Número estimado de clústeres: {n_clusters_estimado}')
plt.show()
```

## En esta implementación:

- Se generan datos de ejemplo utilizando la función make_blobs de scikit-learn.
- Se crea una instancia del modelo MeanShift.
- El modelo se ajusta a los datos utilizando el método fit.
- Se obtienen las etiquetas de los clústeres para cada punto de datos y las coordenadas de los centroides.
- Se grafican los datos y los clústeres utilizando matplotlib.
- El algoritmo Mean-Shift es especialmente útil cuando no conocemos el número de clústeres de antemano y queremos que el algoritmo lo estime automáticamente en función de la densidad de los datos.
