```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Cargar el dataset heart
dt_heart = pd.read_csv('/home/jhonfypy/CursoML_Scikit_learni_jf/heart.csv')
print(dt_heart.head(10))
print('='*50)
print(dt_heart['target'].describe())

# Preparar los datos
X = dt_heart.drop(['target'], axis=1)
y = dt_heart['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)

# Modelo KNeighborsClassifier
knn_class = KNeighborsClassifier().fit(X_train, y_train)
knn_pred = knn_class.predict(X_test)
print('='*68)
print('Accuracy del modelo KNeighborsClassifier:', accuracy_score(knn_pred, y_test))

# Modelo BaggingClassifier con KNeighborsClassifier como estimador base
bag_class = BaggingClassifier(base_estimator=KNeighborsClassifier(), n_estimators=50)
bag_class.fit(X_train, y_train)
bag_pred = bag_class.predict(X_test)
print('='*68)
print('Accuracy del modelo BaggingClassifier:', accuracy_score(bag_pred, y_test))

# Visualizacin de las clasificaciones
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_test['age'], X_test['chol'], c=knn_pred, cmap='viridis')
plt.title('Clasificacin con KNeighborsClassifier')
plt.xlabel('Age')
plt.ylabel('Cholesterol')

plt.subplot(1, 2, 2)
plt.scatter(X_test['age'], X_test['chol'], c=bag_pred, cmap='viridis')
plt.title('Clasificacin con BaggingClassifier')
plt.xlabel('Age')
plt.ylabel('Cholesterol')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Cargar el dataset
dt_heart = pd.read_csv('/home/jhonfypy/CursoML_Scikit_learni_jf/heart.csv')
X = dt_heart.drop(['target'], axis=1)
y = dt_heart['target']

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)

# Lista de clasificadores
classifiers = [
    ('RandomForest', RandomForestClassifier()),
    ('GradientBoosting', GradientBoostingClassifier()),
    ('AdaBoost', AdaBoostClassifier()),
    ('GaussianNB', GaussianNB())
]

# Entrenar y evaluar cada clasificador
results = []
for name, classifier in classifiers:
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results.append((name, accuracy))

# Mostrar resultados
for name, accuracy in results:
    print(f'{name}: Accuracy = {accuracy:.4f}')

# Graficar resultados
plt.figure(figsize=(10, 6))
names, accuracies = zip(*results)
plt.bar(names, accuracies, color='skyblue')
plt.title('Comparacin de Precisin de Clasificadores')
plt.xlabel('Clasificador')
plt.ylabel('Precisin')
plt.ylim(0, 1.0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB

# Cargar el conjunto de datos
dt_heart = pd.read_csv('/home/jhonfypy/CursoML_Scikit_learni_jf/heart.csv')
print(dt_heart.head(10))
print('='*50)
print(dt_heart['target'].describe())

# Dividir los datos en caractersticas (X) y etiquetas (y)
X = dt_heart.drop(['target'], axis=1)
y = dt_heart['target']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)

# Lista de clasificadores
classifiers = [
    ("KNeighborsClassifier", KNeighborsClassifier()),
    ("BaggingClassifier", BaggingClassifier(base_estimator=KNeighborsClassifier(), n_estimators=50)),
    ("LinearSVC", LinearSVC()),
    ("SVC", SVC()),
    ("RandomForestClassifier", RandomForestClassifier()),
    ("GradientBoostingClassifier", GradientBoostingClassifier()),
    ("AdaBoostClassifier", AdaBoostClassifier()),
    ("GaussianNB", GaussianNB())
]

# Entrenar y evaluar cada clasificador
for name, classifier in classifiers:
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_pred, y_test)
    
    print('='*68)
    print(name)
    print(f"Accuracy: {accuracy}")

# Grficos de los resultados
fig, ax = plt.subplots(figsize=(10, 6))
names = [name for name, _ in classifiers]
scores = [accuracy_score(classifier.predict(X_test), y_test) for _, classifier in classifiers]
ax.barh(names, scores)
ax.set_xlabel('Accuracy')
ax.set_title('Accuracy of Different Classifiers')
plt.show()
```
